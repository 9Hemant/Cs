```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(df["v2"],df["v1"], 
                                   test_size=0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
x=CountVectorizer()
```

```python
x_train=x.fit_transform(x_train)
x_test=x.transform(x_test)
```

```python
x.get_feature_names_out()
```

```python
x.vocabulary_
```

```python
from sklearn.svm import SVC
classifier = SVC()
classifier.fit(x_train, y_train)
y=classifier.predict(x_test)
y1=classifier.predict(x_train)
```

```python
print("Training Accuracy:",classifier.score(x_train,y_train))
```

```python
print("Testing Accuracy:",classifier.score(x_test,y_test))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_train,y1)
```

```python
print("Confusion Matrix for Testing Dataset:\n",confusion_matrix(y_test,y))
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
m=MultinomialNB()
```

```python
m.fit(x_train,y_train)
```

```python
z1=m.predict(x_test)
z=m.predict(x_train)
```

```python
print("Testing Accuracy",m.score(x_train,y_train))
```

```python
print("Testing Accuracy:",m.score(x_test,y_test))
```
